```python
import pandas as pd
import os # <-- Permet d'interagir avec les fichiers de l'ordinateur

region_code = 'PACA'
departements = ['04', '05', '06', '13', '83', '84']

```

```python
fname = f'dpe_{region_code}.csv'


# 1. On vérifie si le fichier final existe déjà
if os.path.exists(fname):
    print(f"✅ Le fichier {fname} existe déjà ! Chargement direct...")
    df_region = pd.read_csv(fname)
    print(f"Taille du dataset chargé : {len(df_region)} logements.")

# 2. S'il n'existe pas, on fait la fusion
else:
    print(f"🔄 Le fichier {fname} n'existe pas. Début de la fusion...")
    df_region = pd.DataFrame()
    
    for dep in departements:
        file = f'DPE_dep_{dep}.csv'
        
        # Sécurité : on vérifie que le fichier départemental existe bien
        if os.path.exists(file):
            df = pd.read_csv(file)
            print(f"   -> Ajout du département {dep} : {len(df)} logements")
            df_region = pd.concat([df_region, df], ignore_index=True)
        else:
            print(f"   -> ⚠️ Attention : le fichier {file} est introuvable, il sera ignoré.")
            
    # Sauvegarde du nouveau fichier
    if not df_region.empty:
        df_region.to_csv(fname, index=False)
        print(f"✅ Fusion terminée ! Fichier {fname} sauvegardé avec {len(df_region)} logements.")
    else:
        print("❌ Erreur : Aucun fichier départemental n'a pu être chargé.")
```
